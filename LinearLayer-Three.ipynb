```python
import torch
import torch.nn as nn

# Define your desired values for the weight matrix and bias
custom_weight = torch.tensor([[2.0],[1.0]]).reshape(1, 2)  # Reshape to 1x2 weight matrix
custom_bias = torch.tensor([2.0])  # Bias values (1D tensor)

# Create the linear layer
layer = nn.Linear(2, 1)  # Input size 2, output size 1

# Set the weight and bias manually
with torch.no_grad():
    layer.weight.copy_(custom_weight)
    layer.bias.copy_(custom_bias)

# Now you can use 'layer' with your custom values
```

```python
# Create an input tensor
input_tensor = torch.tensor([[2.0, 3.0]])  # 1D, 2x1 input
```

```python
# Calculate the output
output = layer(input_tensor)
print(f"y={output}")
```

```python
# Exercise A
# Define your desired values for the weight matrix and bias
custom_weight = torch.tensor([[1.0, 1.0],[1.0, -1.0]]).reshape(2, 2)  # Reshape to 2x2 weight matrix
custom_bias = torch.tensor([0.0, 0.0])  # 2x1 Bias values (1D tensor)

# Create the linear layer
layer = nn.Linear(2, 2)  # Input size 2, output size 2

# Set the weight and bias manually
with torch.no_grad():
    layer.weight.copy_(custom_weight)
    layer.bias.copy_(custom_bias)

# Create an input tensor
input_tensor = torch.tensor([[2.0, -1.0]])  # 2x1 input
# Calculate the output
output = layer(input_tensor)
print(f"y={output}")
```
